```python
import dash
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input, Output
import plotly.graph_objs as go
import pandas as pd


data_agricola = pd.read_excel('data_final_agr.xlsx')
list_ciudad = list(data_agricola['ciudad'].unique())
agricultura_4= pd.read_excel('agricultura_4.xlsx')
agricultura_4["Fecha de Inicio del Contrato"] = pd.to_datetime(agricultura_4["Fecha de Inicio del Contrato"], errors='coerce')

#CÁLCULO DE FRECUENCIAS POR LOCALIZACIÓN
frec_localizacion = pd.DataFrame(pd.value_counts(data_agricola["ciudad"]))
frec_localizacion.columns = ["Frec_abs"]
frec_localizacion["Frec_relativa"]=100*frec_localizacion["Frec_abs"]/len(data_agricola)
frec_localizacion =pd.DataFrame(frec_localizacion)
frec_localizacion.reset_index(inplace= True)
#frec_localizacion.reset_index(level=0, inplace= True)
#frec_localizacion

#CÁLCULO DE FRECUENCIAS POR DESTINO DEL GASTO (tipo contrato)
frectipocontrato = pd.DataFrame(pd.value_counts(data_agricola["destino_gasto"]))
frectipocontrato.columns = ["Frec_abs"]
frectipocontrato["Frec_relativa"]=100*frectipocontrato["Frec_abs"]/len(data_agricola)
frectipocontrato.reset_index(inplace= True)
frectipocontrato

#CÁLCULO DE FRECUENCIAS POR ENTIDAD
frecentidad = pd.value_counts(data_agricola["entidad"])

frecentidad = pd.DataFrame(frecentidad)
frecentidad.columns = ["Frec_abs"]
frecentidad["Frec_relativa"]=100*frecentidad["Frec_abs"]/len(data_agricola)
frecentidad.reset_index(inplace= True)


app = dash.Dash(__name__, meta_tags=[{"name": "viewport", "content": "width=device-width"}])



#################

app.layout = html.Div([
    dcc.Tabs([
        
        
         dcc.Tab(label='APROXIMACIÓN SECTOR AGRICOLA', children=[
            html.Div([
                html.H3('RETOS DEL PAIS', style = {'margin-bottom': '4px', 'color': 'white'}),
                html.P('Colombia es un país con diversidad ecosistémica, con tierras fértiles y vocación agropecuaria, donde por la diversidad de sus climas y zonas de vida permite la producción de diferentes cultivos, tanto para consumo nacional como para exportación.', style = {'margin-bottom': '4px', 'color': 'white'}),
                html.P('El sector agrícola de Colombia tiene grandes retos para competir a nivel internacional. Entre los retos y oportunidades que tiene el país se encuentran:', style = {'margin-bottom': '4px', 'color': 'white'}),
                html.P('- Generación tecnología', style = {'margin-bottom': '4px', 'color': 'white'}),
                html.P('- Desarrollo de políticas', style = {'margin-bottom': '4px', 'color': 'white'}),
                html.P('- Pensar en exportación', style = {'margin-bottom': '4px', 'color': 'white'}),
                html.P('- Identificación del potencia productivo (agropecuario) de cada región', style = {'margin-bottom': '4px', 'color': 'white'}),
                html.H3('ENTIDADES ENCARGADAS DE LA GESTIÓN', style = {'margin-bottom': '4px', 'color': 'white'}), 
                html.P('El desarrollo del sector agrícola del país se puede liderar de manera estratégica y técnica, toda vez que Colombia cuenta con entidades que tienen la capacidad de formular y planear este desarrollo a lo largo del país, entre las entidades mas importantes y con mayor inversión desde el tema de contratación estatal se encuentra:', style = {'margin-bottom': '4px', 'color': 'white'}),
                html.P('- Ministerio de Agricultura', style = {'margin-bottom': '4px', 'color': 'white'}),
                html.P('-	Instituto Colombiano Agropecuario – ICA', style = {'margin-bottom': '4px', 'color': 'white'}),
                html.P('-	Agencia de desarrollo rural', style = {'margin-bottom': '4px', 'color': 'white'}),
                html.P('-	Agencia Nacional de Tierras', style = {'margin-bottom': '4px', 'color': 'white'}),
                html.P('-	Autoridad Nacional de pesca (ANAP)', style = {'margin-bottom': '4px', 'color': 'white'}),
                html.P('-	Algunos Resguardos indígenas ', style = {'margin-bottom': '4px', 'color': 'white'}),
                html.H3('DISTRIBUCIÓN DE RECURSOS', style = {'margin-bottom': '4px', 'color': 'white'}),
                html.P('-	Periodo evaluado: Año 2000 a año 2021', style = {'margin-bottom': '4px', 'color': 'white'}),
                html.P('-	Total contratos firmados: 235', style = {'margin-bottom': '4px', 'color': 'white'}),
                html.P('-	Inversión: $12.685.648.396 Billones de pesos', style = {'margin-bottom': '4px', 'color': 'white'}),
                html.P('-	Participación entidades: ICA 67,23% - Min Agricultura 8,51% - AG. Nal. Tierra 8,09% - AG. Dllo Rural 5,96%  - ANAP 5,96% - Resguardo Kankuamo 4,26%', style = {'margin-bottom': '4px', 'color': 'white'}),
                html.P('-	Destino Recursos: Funcionamiento 53,19% - Inversión 35,32% - No Disponible 11,49%', style = {'margin-bottom': '4px', 'color': 'white'}),
                #html.H3('CONTRATOS SECTOR AGRICULTURA - COLOMBIA', style = {'margin-bottom': '5px', 'color': 'white'}),
                #html.P('probando prueba 1 de 1000')
            ])
        ]
        
        ),
        
        
        
        
        
        
        dcc.Tab(label='CANTIDAD DE CONTRATOS', children=[
            html.Div([
                    html.Label('Ciudad'),
                    dcc.Dropdown(id='selector',
                         options=[{'label':i, 'value':i} for i in agricultura_4['Localización'].unique()],
                         value ='Colombia'
                    )], style={'width':'48%', 'display': 'inline-block'}),

                    html.Div([
                    html.Label('Rango fechas'),
                    dcc.DatePickerRange(id='selector_fecha', start_date=agricultura_4['Fecha de Inicio del Contrato'].min(),end_date=agricultura_4['Fecha de Inicio del Contrato'].max())
                    ], style={'width':'48%','float':'right', 'display':'inline-block'}),
    
            
                    html.Div([
                    dcc.Graph(id='barplot_entidad')
                    ], style={'width':'33%', 'float':'left', 'display':'inline-block'}),
            
            
                    html.Div([
                    dcc.Graph(id='barplot_tipocontrato')
                    ], style={'width':'33%', 'float':'center', 'display':'inline-block'}),
    
            
                    html.Div([
                    dcc.Graph(id='barplot_destino')
                    ], style={'width':'33%', 'float':'right', 'display':'inline-block'}),
                    
                    ]),
        
        
        
        
        
        dcc.Tab(label="INDICADORES", children=[
            html.Div([
                html.Div([
                    html.Div([
                        html.H3('CONTRATOS SECTOR AGRICULTURA - COLOMBIA', style = {'margin-bottom': '5px', 'color': 'white'}),
                    ])
                ], className = "create_container1 four columns", id = "title"),

            ], id = "header", className = "row flex-display", style = {"margin-bottom": "10px"}),


            html.Div([
                html.Div([
                    html.P('Ciudad', className = 'fix_label', style = {'color': 'white'}),
                    dcc.Dropdown(id = 'select_ciudad',
                                 multi = False,
                                 clearable = True,
                                 disabled = False,
                                 style = {'display': True},
                                 value = 'Colombia, Bogotá, Bogotá',
                                 placeholder = 'Select Ciudad',
                                 options = [{'label': c, 'value': c}
                                            for c in list_ciudad], className = 'dcc_compon'),

                    html.P('Seleccionar el año', className = 'fix_label', style = {'color': 'white', 'margin-top': '30px'}),
                    dcc.Slider(id = 'slider_year',
                               included = True,
                               updatemode='drag',
                               tooltip={'always_visible': True},
                               min = 2000,
                               max = 2021,
                               step = 1,
                               value = 2000,
                               marks = {str(yr): str(yr) for yr in range(2000, 2022, 5)},
                               className = 'dcc_compon'),

                    html.P('Nombre de la Entidad', className = 'fix_label', style = {'color': 'white', 'margin-top': '30px'}),
                    dcc.Checklist(id = 'radio_items',
                                  options = [{'label': d, 'value': d} for d in sorted(data_agricola['entidad'].unique())],
                                  value=["ICA_NAL"],
                                  style = {'color': 'white'}, className = 'dcc_compon'),

                    ], className = "create_container2 four columns"),


                html.Div([
                    dcc.Graph(id = 'bubble_chart',
                              config = {'displayModeBar': 'hover'}),

                ], className = "create_container2 eight columns"),

            ], className = "row flex-display"),
        ])
    ])
    
    
    
], id= "mainContainer", style={"display": "flex", "flex-direction": "column"})


@app.callback(Output('bubble_chart', 'figure'),
              [Input('select_ciudad', 'value')],
              [Input('slider_year', 'value')],
              [Input('radio_items', 'value')])
def update_graph(select_ciudad, slider_year, radio_items):
    data_agricola1 = data_agricola.groupby(['entidad', 'ciudad', 'destino_gasto', 'año', 'poblacion', 'No_contratos/100k pop'])['No_contratos'].sum().reset_index()
    data_agricola2 = data_agricola1[(data_agricola1['ciudad'] == select_ciudad) & (data_agricola1['año'] >= slider_year) & (data_agricola1['entidad'].isin(radio_items))]

    return {
        'data':[go.Scatter(
                    x=data_agricola2['año'],
                    y=data_agricola2['No_contratos'],
                    text = data_agricola2['destino_gasto'],
                    textposition = 'top center',
                    mode = 'markers + text',
                    marker = dict(
                        size = data_agricola2['No_contratos'] / 10,
                        color = data_agricola2['No_contratos'],
                        colorscale = 'HSV',
                        showscale = False,
                        line = dict(
                            color = 'MediumPurple',
                            width = 2
                        )),
                    hoverinfo='text',
                    hovertext=
                    '<b>ciudad</b>: ' + data_agricola2['ciudad'].astype(str) + '<br>' +
                    '<b>entidad</b>: ' + data_agricola2['entidad'].astype(str) + '<br>' +
                    '<b>destino_gasto</b>: ' + data_agricola2['destino_gasto'].astype(str) + '<br>' +
                    '<b>año</b>: ' + data_agricola2['año'].astype(str) + '<br>' +
                    '<b>poblacion</b>: ' + [f'{x:,.0f}' for x in data_agricola2['poblacion']] + '<br>' +
                    '<b>No_contratos/100k pop</b>: ' + [f'{x:,.0f}' for x in data_agricola2['No_contratos/100k pop']] + '<br>' +
                    '<b>contratos</b>: ' + [f'{x:,.0f}' for x in data_agricola2['No_contratos']] + '<br>'


              )],


        'layout': go.Layout(
             plot_bgcolor='#010915',
             paper_bgcolor='#010915',
             title={
                'text': 'Contratos (fun: funcionamiento - Inv: Inversión - ND: No Disponible): ' + (select_ciudad),

                'y': 0.96,
                'x': 0.5,
                'xanchor': 'center',
                'yanchor': 'top'},
             titlefont={
                        'color': 'gray',
                        'size': 20},

             hovermode='x',
             xaxis=dict(title='<b>Año</b>',
                        tick0=0,
                        dtick=1,
                        color='silver',
                        showline=True,
                        showgrid=False,
                        linecolor='white',
                        linewidth=1,


                ),

             yaxis=dict(title='<b>No. de contratos</b>',
                        color='silver',
                        showline=False,
                        showgrid=True,
                        linecolor='white',

                ),

            legend = {
                'orientation': 'h',
                'bgcolor': '#010915',
                'x': 0.5,
                'y': 1.25,
                'xanchor': 'center',
                'yanchor': 'top'},
            font = dict(
                family = "sans-serif",
                size = 12,
                color = 'white',


                 )
        )

    }

#CREACION DE GRAFICOS Y GROUPBY
@app.callback(Output('barplot_entidad', 'figure'),
              [Input('selector_fecha', 'start_date'), Input('selector_fecha', 'end_date'), Input('selector', 'value')])
def actualizar (fecha_min, fecha_max, selector):
    filtered_df1 = agricultura_4[(agricultura_4["Fecha de Inicio del Contrato"]>=fecha_min)]
    filtered_df1=filtered_df1[(filtered_df1["Fecha de Inicio del Contrato"]<=fecha_max)]
    filtered_df=filtered_df1[(filtered_df1['Localización'] == selector)]
    df_agrupado = filtered_df.groupby("Nombre Entidad")["Ciudad"].agg("count").to_frame(name="cant_contratos").reset_index()
            
    return {
        'data':[go.Bar(x=df_agrupado['Nombre Entidad'],
                       y=df_agrupado['cant_contratos']
                       )],                            
        'layout': go.Layout(
            title = "Contratos por Entidad",
            xaxis={'title': "Nombre Entidad"},
            yaxis={'title': "No. Total Contratos"},
            hovermode='closest'
         )}
               
@app.callback(Output('barplot_tipocontrato', 'figure'),
              [Input('selector_fecha', 'start_date'), Input('selector_fecha', 'end_date'), Input('selector', 'value')])
def actualizar(fecha_min, fecha_max, selector):
    filtered_df1 = agricultura_4[(agricultura_4["Fecha de Inicio del Contrato"]>=fecha_min)]
    filtered_df1=filtered_df1[(filtered_df1["Fecha de Inicio del Contrato"]<=fecha_max)]
    filtered_df=filtered_df1[(filtered_df1['Localización'] == selector)]
    df_agrupado = filtered_df.groupby("Tipo de Contrato")["Ciudad"].agg("count").to_frame(name="cant_contratos").reset_index()
            
    return {
        'data':[go.Bar(x=df_agrupado['Tipo de Contrato'],
                       y=df_agrupado['cant_contratos']
                       )],                            
        'layout': go.Layout(
            title = "Contratos por Tipo de Contrato",
            xaxis={'title': "Tipo de Contrato"},
            yaxis={'title': "No. Total Contratos"},
            hovermode='closest'
            )}
    

@app.callback(Output('barplot_destino', 'figure'),
              [Input('selector_fecha', 'start_date'), Input('selector_fecha', 'end_date'), Input('selector', 'value')])
def actualizar(fecha_min, fecha_max, selector):           
               
    filtered_df1 = agricultura_4[(agricultura_4["Fecha de Inicio del Contrato"]>=fecha_min)]
    filtered_df1=filtered_df1[(filtered_df1["Fecha de Inicio del Contrato"]<=fecha_max)]
    filtered_df=filtered_df1[(filtered_df1['Localización'] == selector)]   
    df_agrupado = filtered_df.groupby("Destino Gasto")["Ciudad"].agg("count").to_frame(name="cant_contratos").reset_index()
            
    return {
        'data':[go.Bar(x=df_agrupado['Destino Gasto'],
                       y=df_agrupado['cant_contratos']
                       )],                            
        'layout': go.Layout(
            title = "Contratos por Destino Gasto",
            xaxis={'title': "Destino Gasto"},
            yaxis={'title': "No. Total Contratos"},
            hovermode='closest'
            )}   
            

if __name__ == '__main__':
    app.run_server()
```
